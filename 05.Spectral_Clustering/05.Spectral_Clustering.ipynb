```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
X = dataset.iloc[:, 3:5].values
X
```

```python
from sklearn.cluster import SpectralClustering
clustering = SpectralClustering(n_clusters=8, eigen_solver=None, n_components=None, 
                                random_state=None, n_init=10, gamma=1.0, affinity='rbf', 
                                n_neighbors=10)
```

```python
label = clustering.fit_predict(X)
```

```python
supervised=pd.DataFrame(dataset)
supervised['Cluster_group']=label
supervised
```

```python
supervised.to_csv("Spectral_Clustering.csv",index=False)
```

```python
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], 
                   fit_reg=False, legend=True, legend_out=True)
```
